```python
#As a football fan, the question is not 'if' our team will turn the ball over.  It is 'when'.  
#To shed some light on that question, I did a quick Survival Analysis using a Weibull Distribution.
import predictr
from predictr import Analysis
import pandas as pd
```

```python
#I pulled data from ryurko's dataset: https://github.com/ryurko/nflscrapR-data
#Specifically I isolated samples of interceptions or fumbles, and listed the time of the game
#that they occured (in seconds).  
nfl = pd.read_csv(r"C:\Users\danbr\OneDrive\Documents\Data Science\NFL Turnover samples time.csv")
```

```python
#Here is the 20 unit sample set.  Bear  in mind there are 3600 seconds in a game.
print(nfl)
```

```python
#Okay, let's set the analysis, where df is the time of turnover, and the bounds set to 
#Parametric Bootrstrap 
x = Analysis(df=[789, 888, 1799, 1974, 2088, 3537, 1131, 1183, 2175, 2985, 3308, 
                         1162, 1257, 1750, 1761, 2588, 2597, 3298, 2131, 2472], show=True, bounds='pbb')
#And run the Means Rank Regression
x.mrr()
```

```python
#Run the Max Likelihood Estimation
x.mle()
```
